```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# arithmetic op.

img1 = cv.imread("sc1.jpg", cv.IMREAD_GRAYSCALE)
img2 = cv.imread("sc2.jpg", cv.IMREAD_GRAYSCALE)
arith_arr = {}
arith_arr["add"] = (img1+img2)
arith_arr["sub"]= (img1-img2)
arith_arr["mult"]=(img1*img2)
arith_arr["div"] = (img1/img2)

for k, v in arith_arr.items():
    plt.figure()
    plt.title(k)
    plt.imshow(v,"gray")
    plt.axis('off')
plt.show()
```

```python
# logical op

from matplotlib.colors import Colormap


img1_gray = cv.imread("sc1.jpg", cv.IMREAD_GRAYSCALE)
img2_gray = cv.imread("sc2.jpg", cv.IMREAD_GRAYSCALE)

_, img1_bin = cv.threshold(img1_gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
_, img2_bin = cv.threshold(img2_gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

plt.title("img1_binary")
plt.imshow(img1_bin,"gray")
plt.show()

plt.title("img2_binary")
plt.imshow(img2_bin,"gray")
plt.show()

img_and = cv.bitwise_and(img1_bin, img2_bin)
plt.title("img_and")
plt.imshow(img_and, "gray")
plt.show()

img_or = cv.bitwise_or(img1_bin, img2_bin)
plt.title("img_or")
plt.imshow(img_or, "gray")
plt.show()


img_xor = cv.bitwise_xor(img1_bin, img2_bin)
plt.title("img_xor")
plt.imshow(img_xor, "gray")
plt.show()

img1_not = cv.bitwise_not(img1_bin)
plt.title("img1_not")
plt.imshow(img1_not, "gray")
plt.show()
```
