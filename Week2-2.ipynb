```python
# morph dilation/ erosion / opening / closing
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread("./istockphoto-816752606-612x612.jpg", cv.IMREAD_GRAYSCALE)

_, img_bin = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

# ()   
kernel = cv.getStructuringElement(cv.MORPH_CROSS, (9, 9))

#  
plt.figure()
plt.title('Disk Kernel (9x9)')
plt.imshow(kernel, cmap='gray')
plt.axis('off')
plt.show()

dilated = cv.dilate(img_bin, kernel)

eroded = cv.erode(img_bin, kernel)

opened = cv.morphologyEx(img_bin, cv.MORPH_OPEN, kernel)

closed = cv.morphologyEx(img_bin, cv.MORPH_CLOSE, kernel)

#  
titles = ['Original', 'Dilation', 'Erosion', 'Opening', 'Closing']
images = [img_bin, dilated, eroded, opened, closed]
plt.figure(figsize=(15,3))
for i in range(5):
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
    plt.show()
```
